```python
Col_Categorical_Dict={}
for col in ObjCol:
    arr = mydata[col].unique()
    dict = {}
    x=0
    for val in arr:
        if isinstance(val, int):
            dict[val]=-1
        else:
            dict[val]=x
            x+=1
    Col_Categorical_Dict[col]=dict
    mydata[col].replace(dict, inplace=True) # Replacing labels with number
```
